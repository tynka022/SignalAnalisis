```python
import os
import sys

#data science
import pandas as pd
import numpy as np

import neurokit2 as nk
from pathlib import Path
```

a) 1000 interwałów

```python
data_1_10 = pd.read_csv("chf201.txt_N.csv", nrows=1000)
rri_1_10 = data_1_10['758'].to_numpy()
```

```python
peaks_1_10, info_1_10 = nk.ecg_peaks(rri_1_10, sampling_rate=100)
hrv_1_10 = nk.hrv_time(peaks_1_10, sampling_rate=100, show=False)
print(hrv_1_10.HRV_SDNN)
```

```python
cleaned_1_10 = nk.signal_detrend(rri_1_10)
peaks_1_10_d, info_1_10_d = nk.ecg_peaks(cleaned_1_10, sampling_rate=100)
hrv_1_10_d = nk.hrv_time(peaks_1_10_d, sampling_rate=100, show=False)
print(hrv_1_10_d.HRV_SDNN)
```

```python
plot_1_10 = nk.signal_plot([rri_1_10, cleaned_1_10], labels=(['dane', 'dane bez trendu']))
```
