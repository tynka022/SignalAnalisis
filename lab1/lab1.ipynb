```python
import os
import sys

#data science
import pandas as pd
import numpy as np

import neurokit2 as nk
from pathlib import Path
```

a) 1000 interwałów

```python
data_1_10 = pd.read_csv("chf201.txt_N.csv", nrows=1000, header = None)
rri_1_10 = data_1_10[0].to_numpy()
```

```python
std_1_10= np.std(rri_1_10)
std_1_10
```

```python
cleaned_1_10 = nk.signal_detrend(rri_1_10)
std_1_10c = np.std(cleaned_1_10)
std_1_10c
```

```python
plot_1_10 = nk.signal_plot([rri_1_10, cleaned_1_10], labels=(['dane', 'dane bez trendu']))
```
