```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
from cohen import *
```

```python
fs = 500 
N = 512
f1 = 20
f2 = 100
tn = np.arange(1,N)/fs
w=N/fs
sygnal = sig.chirp(tn,f1, w, f2)
```

```python
fig, ax = plt.subplots()
ax.plot(sygnal)
plt.show
```

```python
nff =16
f, t, Sxx = sig.spectrogram(sygnal, fs=fs, nperseg=nff//2, nfft=nff)
plt.pcolormesh(t, f, Sxx)
plt.xlabel('czas [s]')
plt.ylabel('czstotliwo [Hz]')
plt.title('Spektrogram sygnau')
plt.colorbar()
plt.show()
```
