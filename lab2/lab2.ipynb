```python
import numpy as np
import scipy.signal as sig
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
fs = 48000
f = np.fft.rfftfreq(2048, 1/fs)
```

```python
n = np.arange(2048)
```

```python
sinus = np.sin(2*np.pi*n*1000/fs)
widmo = np.fft.fft(sinus)
widmo_amp = np.abs(np.fft.rfft(sinus))

fig, ax = plt.subplots()
ax.plot(n,sinus , label="sinus")
ax.legend()
plt.show
```

```python
fig, ax = plt.subplots()
ax.plot(f, widmo_amp, label = "widmo")
ax.legend()
plt.show
```

```python
f1=200
l = np.arange(150, 450, 1)
prostokat = ((((l*2*f1/fs+1)%2)-1)>=0).astype(int)
plt.plot(l,prostokat, label = "prostokat")
```

```python
plt.plot(l, (prostokat * sinus[150:450]), label='prostokat')
```

```python
plt.plot(l, (np.abs(np.fft.fft(prostokat*sinus[150:450]))), label='widmo prostokata')
```
