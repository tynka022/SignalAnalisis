```python
import numpy as np
import scipy.signal as sig
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
#Zad1
# częstotliwość próbkowania
fs = 48000
```

```python
#a) sygnał nieskończony
n = np.arange(2048)
sinus = np.sin(2*np.pi*n*1000/fs)

fig, ax = plt.subplots()
ax.plot(n[0:800],sinus[0:800])
plt.show
```

```python
#widmo sygnału nieskończonego
f = np.fft.rfftfreq(2048, 1/fs)
widmo_amp = np.abs(np.fft.rfft(sinus))/ 1024

fig, ax = plt.subplots()
ax.plot(f, widmo_amp)
plt.show
```

```python
#b) sygnał okna prostokątnego
l = np.linspace(0, 1, 2048, endpoint=False)
prostokat = sig.square(l*2*np.pi+150)+1
fig, ax = plt.subplots()
ax.plot(prostokat)
plt.show
```

```python
#widmo sygnału okna prostokątnego
wid_pr =  np.abs(np.fft.rfft(prostokat))/ 1024

fig, ax = plt.subplots()
ax.plot(f, wid_pr)
plt.show
```

```python
#c) sygnał skończony
sygn = prostokat*sinus
fig, ax = plt.subplots()
ax.plot(sygn)
plt.show
```

```python
#widmo sygnału skończonego
wid_s =  np.abs(np.fft.rfft(sygn))/ 1024

fig, ax = plt.subplots()
ax.plot(f, wid_s)
plt.show
```

```python
#d) sygnał okna nieprostokatnego
niepr = np.blackman(2048)

fig, ax = plt.subplots()
ax.plot(niepr)
plt.show
```

```python
#widmo okna nieprostokatnego
wid_niepr =  np.abs(np.fft.rfft(niepr))/ 1024

fig, ax = plt.subplots()
ax.plot(wid_niepr)
plt.show
```

```python
#e) sygnał skończony wygładzonego okna
syg_no = np.blackman(2048) * sinus

fig, ax = plt.subplots()
ax.plot(syg_no)
plt.show
```

```python
#widmo sygnału skończonego wygładzonego okna
wid_no = np.abs(np.fft.rfft(np.blackman(2048) * sinus))/ 1024

fig, ax = plt.subplots()
ax.plot(wid_no)
plt.show
```

```python
#Zad2
#synał
rate, sygnal = wavfile.read('chord.wav')
#częstotliwość Nyquista
fNyq=rate/2
print(fNyq)

fig, ax = plt.subplots()
ax.plot(sygnal)
plt.show
```

```python
#widmo sygnału
widmo = np.abs(np.fft.rfft(sygnal))/rate
f = np.fft.rfftfreq(len(sygnal), 1/rate)

fig, ax = plt.subplots()
ax.plot(f[0:10000], widmo[0:10000])
plt.show
```

```python
#decymacja sygnału
sygnal1 = sig.decimate(sygnal,2)

fig, ax = plt.subplots()
ax.plot(sygnal1)
plt.show

#widmo sygnału decymowanego
wid_dec=np.abs(np.fft.rfft(sygnal1)) /rate
f = np.fft.rfftfreq(len(sygnal1), 1/rate)

fig, ax = plt.subplots()
ax.plot(f[0:10000], wid_dec[0:10000])
plt.show
```

```python
#ponowna decymacja sygnału
sygnal2 = sig.decimate(sygnal1,2)

fig, ax = plt.subplots()
ax.plot(sygnal2)
plt.show

#widmo sygnału decymowanego
wid_dec2=np.abs(np.fft.rfft(sygnal2)) /rate
f = np.fft.rfftfreq(len(sygnal2), 1/rate)

fig, ax = plt.subplots()
ax.plot(f[0:10000], wid_dec2[0:10000])
plt.show
```

```python
#ponowna decymacja sygnału
sygnal3 = sig.decimate(sygnal2,2)

fig, ax = plt.subplots()
ax.plot(sygnal3)
plt.show

#widmo sygnału decymowanego
wid_dec3=np.abs(np.fft.rfft(sygnal3)) /rate
f = np.fft.rfftfreq(len(sygnal3), 1/rate)

fig, ax = plt.subplots()
ax.plot(f[0:10000], wid_dec3[0:10000])
plt.show
```
